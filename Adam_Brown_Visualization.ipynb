```python
def find_outlier(string):
    
    # read csv dataframe to use for manipulation, csv was made from videos in module
    df_csv = pd.read_csv('df_export.csv')
    
    #get row values and store in numpy array
    df_value = df_csv[string].to_numpy()
    
    # initialize arrays to hold data we will grab
    outliers = []
    country = []
    countries = df_csv['Country'].tolist()
    
    # calculate mean, std, and deviation
    mean = df_value.mean()
    std = df_value.std()
    deviation = 1.8 * std
    
    # count for country index as we go through deviated values
    count = 0
    
    # for every element in row specifed check if it is greater than sum of mean and deviation or less than
    # difference of mean and deviation
    # add it to the country array, which states the index and value countries 
    # that are associated with that deviation
    for i in df_value:
        if i > (mean + deviation) or i < (mean - deviation):
            country.append(countries[count])
        count += 1

    # print the outlier countries in that column
    print("The outliers in " + string + " are " + str(country))
            
    
find_outlier('Average')
find_outlier('Math')
find_outlier('Science')
find_outlier('Reading')
```
